# Lemmatization

```python
# Word --> root word

# import nltk
# nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
wordnetlemmatizer = WordNetLemmatizer()
print(f"{'going'}-->{wordnetlemmatizer.lemmatize('going')}")
```

```python
print(f"noun: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='n')}")

print(f"verb: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='v')}")

print(f"{'adjective: {going'}-->{wordnetlemmatizer.lemmatize('going', pos='a')}")

print(f"adverb: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='r')}")

```
