# Lemmatization

```python
# Word --> root word

# import nltk
# nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
wordnetlemmatizer = WordNetLemmatizer()
print(f"{'going'}-->{wordnetlemmatizer.lemmatize('going')}")
```

```python
print(f"noun: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='n')}")

print(f"verb: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='v')}")

print(f"adjective: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='a')}")

print(f"adverb: {'going'}-->{wordnetlemmatizer.lemmatize('going', pos='r')}")
```

```python
words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'program', 'programs', 'history', 'finally', 'finalized']
```

```python
for word in words:
    print(f"noun::{word}-->{wordnetlemmatizer.lemmatize(word, pos='n')}")
    print(f"adjective::{word}-->{wordnetlemmatizer.lemmatize(word, pos='a')}")
    print(f"verb::{word}-->{wordnetlemmatizer.lemmatize(word, pos='v')}")
    print(f"adverb::{word}-->{wordnetlemmatizer.lemmatize(word, pos='r')}")
    print("-"*50)
```
