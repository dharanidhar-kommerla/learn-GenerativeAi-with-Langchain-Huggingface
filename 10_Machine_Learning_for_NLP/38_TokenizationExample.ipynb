```python
## corpus --> documents
## documents --> words

from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

corpus = "Hey. My name is Dharanidhar Kommerla. I am a software Engineer at TCS. My Primary role developing Testsuites using TESTNG framework. I am also Learning GenAI and AgenticAI from krish naik's Udemy course."

# corpus --> documents
sentences = sent_tokenize(corpus)
print(sentences)
print('-'*100)

# documents --> words
for sentence in sentences:
    words = word_tokenize(sentence)
    print(words)
print('-'*100)

# corpus --> words
print(word_tokenize(corpus))

# here fullstops, commas, exclamatory, --> punctuations are treated as a seperate words
```

```python
# TreebankWordDetokenizer
from nltk.tokenize import TreebankWordDetokenizer
tokenizer = TreebankWordDetokenizer() # creating the object for this class
words = tokenizer.tokenize(corpus)
words 
# here fullstop is not treated as seperateword --> here punctuations are not treated as a sperate words
# only last fullstop is treated as seperateword
```
