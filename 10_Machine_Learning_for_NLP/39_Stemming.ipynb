# Stemming

```python
# Stemming
# word --> word stem
# eating, eaten, eat --> eat
# going, gone, goes --> goes
```

```python
words = ['writing', 'writes', 'written', 'going', 'gone', 'goes', 'eating', 'eaten', 'eat', 'programmer', 'programming', 'programmed', 'history', 'final', 'finally', 'finalized', ]
```

# PorterStemmer

```python
# PorterStemmer
from nltk.stem import PorterStemmer
porterstemmer = PorterStemmer()
for word in words:
    print(f"{word} --- {porterstemmer.stem(word)}")
    # history --- histori # for some of the words we don't get excat rootwords
```

```python
porterstemmer.stem('congratulations')
```

```python
porterstemmer.stem("sitting")
```

# RegexpStemmer class

```python
# RegexpStemmer
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|es$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('goes')
```

# Snowball Stemmer

```python
## snowball stemmer --> better than porterstemmer

from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(f'{word} --> {snowballstemmer.stem(word)}')
```

```python
# still for word 'history' --> 'histori'
```

```python
print(f"porterstemmer -- {'sportingly'} --- {porterstemmer.stem('sportingly')}")
print(f"porterstemmer -- {'fairly'} --- {porterstemmer.stem('fairly')}")
print("\n")
print(f"snowballstemmer -- {'sportingly'} --- {snowballstemmer.stem('sportingly')}")
print(f"snowballstemmer -- {'fairly'} --- {snowballstemmer.stem('fairly')}")
# better result with snowballstemmer
```

```python
# still stemming have some limitations
print(f"snowballstemmer -- {'goes'} --- {snowballstemmer.stem('goes')}")

# for chatbot application -- we use lematization techniques -- which solves problems of stemming
```
